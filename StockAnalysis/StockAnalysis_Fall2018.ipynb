## - Five assets picked for analysis


### BTC-USD (Currency Pair) - Bitcoin / US Dollar
This is a currency pair for BTC (base) and USD (quote). Here, Bitcoin is being compared against the US Dollar, the counter currency. It shows us how many US Dollars must be purchased for one Bitcoin. Bitcoin is an open-source, cryptocurrency that operates in a decentralized environment with no central authority or banks. All transactions and flow of Bitcoin are carried out collectively by the network.  


### GLD (ETF) - SPDR Gold Shares
This is an exchange-traded fund that tracks the price of gold. It is part of the family of ETFs managed by the company State Street Global Advisors. This is one of the top ten biggest holders of gold in the world.


### SVXY (ETF) - ProShares Short VIX Short-Term Futures
This is an exchange-traded fund provided by ProShares that provides inverse exposure to the VIX Short-Term Futures Index by taking the first two months of VIX future positions and creating a weighted average maturity of one month. It offers a daily one-half inverse (-0.5x) exposure to short-term VIX futures in a liquid commodities pool, as no product can directly provide exposure to the index itself. VIX is the CBOE Volatility Index and is a measure of the future-expected volatility of 30 days using the price inputs of the S&P 500 Index. It is also known as the “Fear Index” and is a measure of the market risk and investor sentiment. 


### ^FTSE (Index) - The Financial Times Stock Exchange 100 Index
This is a share index, colloquially known as the “Footsie”, that tracks the top 100 companies listed on the London Stock Exchange by market capitalization. It is co-owned by The Financial Times and the London Stock Exchange. It is the most well-known index from the FTSE. It is often considered an indicator of the health of UK’s companies and economy in general representing around 80% of the entire London Stock Exchange market capitalization, although the FTSE 250, the lesser known index, is a better indicator as it comprises a smaller proportion of international companies.  



### MGGPX (Mutual Fund) - Morgan Stanley Institutional Fund, Inc. Global Opportunity Portfolio Class A
This is a Class A mutual fund managed by the company Morgan Stanley. The top sectors included in this fund, according to Morning Star as of 9/30/18, are Consumer Cyclical at 29.78%, Technology at 24.99%, Consumer Defensive at 18.47%, Financial Services at 16.02%, and Industrials at 6.05%.



## - Where and when are they traded? 

### BTC-USD (Currency Pair) 
BTC-USD is a currency pair and is measured by the CCC index, which was created by the company CryptoCompare. BTC-USD is based on cryptocurrency, and therefore can be traded on a number of different cryptocurrency exchange markets. Some of the most well-known exchanges for Bitcoin to USD are Bitfinex, Coinbase, Binance, Kraken, and Bitstamp. Because cryptocurrency is a digital currency and is traded throughout the world, the market is technically open 24/7, 7 days a week. However, it is better to trade when global market activity is high and the cryptocurrency exchange sites themselves may be down for maintenance. 


### GLD (ETF)
GLD is traded on the NYSE Arca. It was previously known as the ArcaEx and is owned by the Intercontinental Exchange. It is an electronic securities exchange where exchange-traded funds (ETFs), exchange-traded notes (ETNs), and exchange-traded vehicles (ETVs) are listed and traded. It is the world’s leading ETF exchange in terms of listings and volume. Investors can trade on the NYSE Arca from 9:30 A.M. to 4 P.M. ET from Monday to Friday. There are also opening session hours from 4 A.M. to 9:30 A.M. ET, pre-opening session hours from 3:30 A.M. to 4 A.M. ET, and extended hours from 4 P.M. to 8 P.M. ET.


### SVXY (ETF)
SVXY is also traded on the NYSE Arca, as GLD is. The trading hours are therefore the same for SVXY as it is for GLD.

### FTSE (Index)
FTSE is an index fund and can be bought by companies that offer index mutual funds and ETFs to investors. Some tracker funds that track the FTSE 100 Index include the LEGAL & GENERAL UK 100 INDEX TRUST CLASS C - ACCUMULATION (GBP) from Legal & General, FTSE 100 UCITS ETF (VUKE) from Vanguard, iShares FTSE 100 Ucits ETF from BlackRock, HSBC FTSE 100 Units ETF from HSBC, and SPDR FTSE UK All Share UCITS ET from State Street Global Advisors, informally known as Spyders or Spiders. Index funds are not traded the way stocks are traded on the stock exchange market. They are a form of passive investing and are Instead bought through companies that offer these funds and are managed by them, similar to the way mutual funds are handled. 

### MGGPX (Mutual Fund)
MGGPX and mutual funds, in general, are not traded on a secondary market (such as the stock exchange market) and is instead a direct transaction between the buyer and the owner of the mutual fund. Mutual funds are only traded once per day (daily) after markets close, which is generally 4 p.m. ET. The execution of clearing of a trade is done over the next one to three business days, depending on the type of fund and the company. Equity and bond funds usually clear within one day while commodities and other types usually clear within three days after the transaction date.

# 6. Regression Anaylsis
- SP500 is independent variable and assets are dependent variables
- Display each regression as a subplot in one figure
- Calculate and explain the intercept, coefficient (slope), R2, and the Mean squared error.



```python
import statsmodels
import statsmodels.formula.api as smf
```

```python
ds=pd.DataFrame()
```

```python
ds['btc']=df_dpc[('Adj Close', 'BTC-USD')]
ds['gld']=df_dpc[('Adj Close', 'GLD')]
ds['svxy']=df_dpc[('Adj Close', 'SVXY')]
ds['ftse']=df_dpc[('Adj Close', '^FTSE')]
ds['mggpx']=df_dpc[('Adj Close', 'MGGPX')]
ds['sp500']=df_dpc[('Adj Close', '^GSPC')]
```

### 1. BTC-USD

```python
model1=smf.ols(formula='btc~sp500',data=ds)
r1=model1.fit()
print(r1.summary())
```

```python
print('MSE:',r1.mse_resid)
```

### BTC-USD Explanation

The intercept represents the expected value of BTC-USD Adj Close, which is 0.7320.

The coefficient (slope) refers to the mean change in the BTC-USD Adj Close for one unit of change in S&P 500 while holding other predictors in the model constant. Therefore, for every additional one percentage increase in BTC-USD Adj Close, we can expect the S&P 500 Adj Close to increase by 0.2135 percent. 

R-squared is also known as the coefficient of determination and is the percentage of the response variable variation that is explained by a linear model. It indicates the percentage of variance in the dependent variable, BTC-USD Adj Close, that the independent variable, S&P 500 Adj Close, explains collectively. 

In this scenario, BTC-USD vs. S&P 500’s model has an r-squared of 0%. This means 0% of the variability in the data around the mean can be explained by S&P 500 Adj Close. This may imply that there is a very weak, non-existent, relationship between BTC-USD Adj Close and S&P 500 Adj Close.

Mean squared error is the average of the square of errors. It tells you how close the regression line is to the set of points. The errors are produced by taking the difference in the distance between the observed values (the data points) and the predicted ones (the regression line). They are then squared to remove any negative values that may cancel out the positive values. The closer the MSE is to 0, the better the model is to finding the line of best fit. 

Here, the mean squared error for BTC-USD vs. S&P 500 is 27.5000080578.

## - Explain your regression findings

Regression findings have shown that different assets have varying relationships with the S&P 500.

This relationship has a R-squared of 0 and the regression line is almost horizontal, which shows that there is very little connection between Bitcoin and the S&P 500, if at all. We assume that this would hold true for other cryptocurrencies similar to Bitcoin as well. 

Using GLD as a representative of gold and in turn precious metals in general, it appears that there is a weak negative correlation between the prices of gold and the S&P 500, with a negative slope and a low R-squared. Because this relationship appears weak with a small negative slope, would not recommend extrapolating the conclusion of a negative correlation to represent other precious metals as well. Using GLD as just a representative for gold, from this data it is possible to conclude the prices of gold moves in counter to the S&P 500. 

Using SVXY as a representative of ETFs of the short-term future variety that tracks the VIX, it appears that there is a considerable relationship between the prices of SVXY and the stocks of S&P 500. With the highest r-squared out of all five models, there is reason to believe that the prices of SVXY and other ETFs of its variety moves in considerably close correspondence to the S&P 500. 

Using FTSE as a representative of the British economy and in turn other countries’ indexes and the interconnectivity of the global market in general, it appears that there is a considerable relationship between the FTSE index and the S&P 500 index. There is a slope of 0.5960 in relation to the S&P 500 which one could use to conclude there is some relation and dependence to the S&P 500. However, it has a rather low R-squared and average MSE compared to the other assets we used. This could potentially be explained by the British economy still having considerable autonomy from the American economy. 

Using MGGPX as a representative of itself, as mutual fund mixtures are usually not disclosed to the public and therefore can be all different, it appears that there is a considerable relationship between the prices of MGGPX and the S&P 500. It has an R-squared of 40.7% and a slope of 0.8465, which is the largest slope out of all five models, which could be taken as meaning there is a strong positive correlation between MGGPX and the S&P 500. From this, there is reason to believe that the portfolio mixture used by Morgan Stanley has noticeable similarities to the S&P 500 and/or moves closely in relation and is affected by the S&P 500.  

One should note, however, that correlation may not always equal causation and was based on sample data. Therefore, these findings and conclusions should not be taken as a definitive truth.
